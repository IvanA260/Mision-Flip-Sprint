````python
# --- CREAR ARCHIVO CAN√ìNICO (WORM) ---
import hashlib

nombre_archivo = f"00_setup_{ALUMNO.replace(' ', '_')}.txt"

# Crea un archivo base reproducible (snapshot)
with open(nombre_archivo, "w", encoding="utf-8") as f:
    f.write(f"Alumno: {ALUMNO}\n")
    f.write(f"Fecha: {FECHA}\n")
    f.write("Secciones: 1. DATOS, 2. M√âTODO, 3. M√âTRICAS, 4. DISCUSI√ìN\n")

# Calcula su hash SHA-256
with open(nombre_archivo, "rb") as f:
    sha = hashlib.sha256(f.read()).hexdigest()

print("Archivo creado:", nombre_archivo)
print("SHA-256 generado:", sha)
````

````python
import json
registro = {
    "archivo": nombre_archivo,
    "sha256": sha,
    "fecha": datetime.datetime.now().isoformat(),
    "autor": ALUMNO
}
with open("registro_hash.json", "w", encoding="utf-8") as f:
    json.dump(registro, f, indent=2, ensure_ascii=False)

print("Registro creado ‚úÖ ‚Üí registro_hash.json")
print(json.dumps(registro, indent=2, ensure_ascii=False))
````

````python
EEE = {
    "Exactitud": "‚úîÔ∏è Resultados correctos y hash v√°lido",
    "Explicabilidad": "‚úîÔ∏è Celdas claras y comentadas",
    "Econom√≠a": "‚úîÔ∏è Notebook limpio y sin redundancia"
}

# Guarda resumen en README.md
with open("README.md", "w", encoding="utf-8") as f:
    f.write("# S1 ‚Äî Plantilla reproducible\n\n")
    f.write(f"- Alumno: {ALUMNO}\n")
    f.write(f"- Archivo: {nombre_archivo}\n")
    f.write(f"- SHA-256: {sha}\n")
    f.write(f"- Fecha: {registro['fecha']}\n\n")
    f.write("## Autoevaluaci√≥n EEE\n")
    for k,v in EEE.items():
        f.write(f"- {k}: {v}\n")

print("README.md creado ‚úÖ")
for k,v in EEE.items():
    print(f"{k}: {v}")
````

````python
# --- COMPROBAR QUE EL ARCHIVO NO CAMBI√ì ---
with open(nombre_archivo, "rb") as f:
    nuevo_hash = hashlib.sha256(f.read()).hexdigest()

print("Hash original :", sha)
print("Hash recalculado:", nuevo_hash)
print("Coinciden ‚úÖ" if nuevo_hash == sha else "Error ‚ùå: el archivo cambi√≥")
````
